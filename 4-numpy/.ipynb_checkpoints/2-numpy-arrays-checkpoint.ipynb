## Creating Arrays

Notice that you have to pass in a list of numbers.

```python
np.array(1,2,3)  ## ERROR: won't work

np.array([1,2,3]) ## Correct
```


````python
# integer array
a1 = np.array([1, 4, 2, 5, 3])
print (a1)
print (type(a1))
````

````python
# 1D array
arr1d = np.array([1, 2, 3, 4, 5])
print(arr1d)

# 2D array
arr2d = np.array([[1, 2, 3], [4, 5, 6]])
print(arr2d)
````

````python
# Create a float array
arrfloat = np.array([1, 2, 3, 4], dtype='float32')
print (arrfloat)
print (type(arrfloat))
print (arrfloat.dtype)
````

````python
# Sequence of numbers with arange
a3 = np.arange(100)
a3
````

## Accessing Numpy Arrays

Zero based index.


````python
a = np.arange(10)
a
````

````python
# first element
a[0]
````

````python
# last elemnt
a[9]
````

````python
# this is an error!
a[10]
````

````python
# Using negative indexes to walk backwards!
# This is very cool, try it out

a[-1]  # last element
````

````python
a [-2]  # second to last element
````

## Array Operations


````python
# Try to multiply sequence by a scalar
a = np.arange(10)
a

a * np.pi
````

````python
# Element-wise operations
arr = np.array([1, 2, 3, 4, 5])
print(arr + 2)    # Add 2 to each element
print(arr * 3)    # Multiply each element by 3

# Array broadcasting
arr1 = np.array([[1, 2, 3], [4, 5, 6]])
arr2 = np.array([10, 20, 30])
print(arr1 + arr2)  # Add arr2 to each row of arr1
````

## Multidimensional Arrays

We can make multi-dimensional arrays from single dimensions with shape.


````python
a = np.array([1,2,3,4,5,6])
a
````

````python
b = a.reshape(2,3)
b
````

````python
c = a.reshape (3,2)
c
````

We can access elements of the 2-D array using array subscription operators.

Remember we do **this**:

```python
b[1,2]
```

not **this**

```python
b[1][2] # BAD!
```
      

````python
b[1,2]
````
