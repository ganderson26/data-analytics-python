**1. Read the unemployment.xlsx file into a DataFrame.**

Read the spreadsheet into a DataFrame and examine the information. Notice any missing and unusual fields.


````python
# Import libraries
import pandas as pd

# Create DataFrame from reading Excel file using pandas
# pandas is using openpyxl depending on the file extension under the hood
df = pd.read_excel('unemployment.xlsx', sheet_name='Rates', usecols='A:M')

# A little debugging
print(df.head(50))

# Looks like we have some NaNs
````

**2. Cleanup the data.**

Check missing values in the month columns and replace with the mean.


````python
# Check for missing values in the month columns
print(df.loc[df['Jan'].isnull() | df['Feb'].isnull() | df['Mar'].isnull() | df['Mar'].isnull() | 
      df['Apr'].isnull() | df['May'].isnull() | df['Jun'].isnull() | df['Jul'].isnull() | 
      df['Aug'].isnull() | df['Sep'].isnull() | df['Oct'].isnull() | df['Nov'].isnull() |
      df['Dec'].isnull()])
````

````python
# Create a function to fill in missing data with mean()

def convert_nan(column):
    df[column] = pd.to_numeric(df[column], errors='coerce')
    df[column].fillna(df[column].mean())
    return df
    
df = convert_nan('Jan')
df = convert_nan('Feb')
df = convert_nan('Mar')
df = convert_nan('Apr')
df = convert_nan('May')
df = convert_nan('Jun')
df = convert_nan('Jul')
df = convert_nan('Aug')
df = convert_nan('Sep')
df = convert_nan('Oct')
df = convert_nan('Nov')
df = convert_nan('Dec')

print(df)
````

**3. Study the data using a Plot.**


````python
# Plot the data
import matplotlib.pyplot as plt

# Get the average of the months
df['Avg'] = df[['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']].mean(axis=1)

# Create a Plot
plt.plot(df['Year'], df['Avg'], color='red', marker='o')
plt.grid(True)
plt.show(block=False)

````

**4. Beautify the Plot with colors and labels.**


````python
# Create a Plot
plt.plot(df['Year'], df['Avg'], color='red', marker='o')
plt.title('Unemployment Rate By Year', fontsize=14)
plt.xlabel('Year', fontsize=14)
plt.ylabel('Rate', fontsize=14)
plt.grid(True)
plt.show(block=False)
````

**5. Save the spreadsheet.**


````python
df.to_excel('updated-unemployment.xlsx')
````
